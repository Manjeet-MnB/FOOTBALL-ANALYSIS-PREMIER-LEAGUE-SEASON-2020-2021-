```python
# Importing the necessary libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Importing the Dataset

df = pd.read_csv("EPL_20_21.csv", encoding = 'latin1')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
# Checking NULL Values

df.isna().sum()
```

```python
# Creating Columns for Mins_Per_Match and Goals_Per_Match

df['Mins_Per_Match'] = (df['Mins'] / df['Matches']).astype(int)
df['Goals_Per_Match'] =  (df['Goals'] / df['Matches']).astype(float)
```

```python
df.head(5)
```

```python
# Calculating Total Goals Scored in the Season

Total_goals = df['Goals'].sum()
print(Total_goals)
```

Total 986 Goals were scored in the last Season of EPL 2020-2021

```python
# Calculating Goals Scored Via Penalties

Total_PenaltyGoals = df['Penalty_Goals'].sum()
print(Total_PenaltyGoals)
```

Total 102 Goals were scored via Penalties

```python
# calculating Penalty Attempts

Total_PenaltyAttempts = df['Penalty_Attempted'].sum()
print(Total_PenaltyAttempts)
```

Total there were 125 penalty Attempts

```python
# Creating Pie chart for penalties scored and missed

plt.figure(figsize=(13,6))
pl_not_scored = df['Penalty_Attempted'].sum() - Total_PenaltyGoals
data = [pl_not_scored, Total_PenaltyGoals]
labels = ['Penalties missed', 'Penalties Scored']
color = sns.color_palette('Set2')
plt.pie(data, labels = labels, colors = color, autopct = '%.0f%%')
plt.show()
```

```python
# Most Players from which country with Bar chart

nationality = df.groupby('Nationality').size().sort_values(ascending = False)
nationality.head(10).plot(kind = 'bar', figsize = (12,6), color = sns.color_palette('magma'))
```

```python
# Clubs with maximum Players in thier squad

df['Club'].value_counts().nlargest(5).plot(kind = 'bar', color = sns.color_palette('viridis'))
```

```python
# Clubs with Least Players in thier squad

df['Club'].value_counts().nsmallest(5).plot(kind = 'bar', color = sns.color_palette('viridis'))
```

```python
# Players based on age group

Under20 = df[df['Age'] <= 20]
age20_25 = df[(df['Age'] > 20) & (df['Age'] <= 25)]
age25_30 = df[(df['Age'] > 25) & (df['Age'] <= 30)]
Above30 = df[df['Age'] > 30]
```

```python
# total under 20 players in each club

Players_under_20 = df[df['Age'] < 20]
Players_under_20['Club'].value_counts().plot(kind = 'bar', color = sns.color_palette('cubehelix'))
```

```python
# Avg Age of players in each club

plt.figure(figsize = (12,6))
sns.boxplot(x = 'Club', y = 'Age', data = df)
plt.xticks(rotation = 90)
```

```python
num_player = df.groupby('Club').size()
data = (df.groupby('Club')['Age'].sum() / num_player)
data.sort_values(ascending = False)
```

```python
# Total Assists from each club

Assits_by_clubs = pd.DataFrame(df.groupby('Club', as_index = False)['Assists'].sum())
sns.set_theme(style = 'whitegrid', color_codes = True)
ax = sns.barplot(x = 'Club', y = 'Assists', data = Assits_by_clubs.sort_values(by = 'Assists'),palette = 'Set2')
ax.set_xlabel('Club', fontsize = 30)
ax.set_ylabel('Assists', fontsize = 20)
plt.xticks(rotation = 90)
plt.rcParams['figure.figsize'] = (20,8)
plt.title('Plot of Clubs vs Total Assists', fontsize = 20)
```

```python
# Top 10 Assists

top_10_assists = df[['Name','Club','Assists','Matches']].nlargest(n =10, columns = 'Assists')
top_10_assists
```

 Harry Kane has maximum assists = 14 

```python
# Total Goals from each club

Goals_by_clubs = pd.DataFrame(df.groupby('Club', as_index = False)['Goals'].sum())
sns.set_theme(style = 'whitegrid', color_codes = True)
ax = sns.barplot(x = 'Club', y = 'Goals', data = Goals_by_clubs.sort_values(by = 'Goals'),palette = 'rocket')
ax.set_xlabel('Club', fontsize = 30)
ax.set_ylabel('Goals', fontsize = 20)
plt.xticks(rotation = 90)
plt.rcParams['figure.figsize'] = (20,8)
plt.title('Plot of Clubs vs Total Goals', fontsize = 20)
```

```python
# Most Goals by players

top_10_Goals = df[['Name','Club','Goals','Matches']].nlargest(n =10, columns = 'Goals')
top_10_Goals
```

 Harry Kane has most number goals = 23

```python
# Pie Chart - Goals with assists and without assists

plt.figure(figsize=(14,7))
assists = df['Assists'].sum()
data = [Total_goals - assists, assists]
labels = ['Goals without assists', 'Goals with assists']
color = sns.color_palette('Set1')
plt.pie(data, labels = labels, colors = color, autopct = '%.0f%%')
plt.show()
```
